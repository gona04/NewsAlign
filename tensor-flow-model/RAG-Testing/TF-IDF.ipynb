```python
#Download some models from nltk 
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
# Sample text 
text = "India Emergency was imposed to bring down morale of the country, says EAM Jaishankar"
```

```python
#Tokenize into sentences 
nltk.sent_tokenize(text)
```

```python
nltk.word_tokenize(text)
```

```python
#Pre-processing text 
def preprocessing(text):
    #Lower case text
    text = text.lower()
    #Tokenization
    tokens = nltk.word_tokenize(text)
    #Removing anything that is not a word or a number
    return [word for word in tokens if word.isalnum()]
```

```python
preprocessing(text)
```

```python
#Apply word tokenizer and lowercase to text
preprocessed_text =[" ".join(preprocessing(doc)) for doc in document]
```

```python
#Creating an instance of vectorizer 
vectorizer = TfidfVectorizer()
```

```python
#Fit and transform the pre-processed docs 
tfidf_matrix = vectorizer.fit_transform(preprocessed_text)

print(f"The shape of TF-IDF matrix is {tfidf_matrix.shape}")
```

```python
query = "Elon musk"
```

```python
def search_tfidf(query, vectorizer, tfidf_matrix):
    query_vec = vectorizer.transform([query])
    similartities = cosine_similarity(tfidf_matrix, query_vec).flatten()
    sorted_similarities = list(enumerate(similartities))
    result = sorted(sorted_similarities, key=lambda x:x[1], reverse=True)
    return result
```

```python
search_similarities = search_tfidf(query, vectorizer, tfidf_matrix)
# print(search_similarities)
for doc_index, score in search_similarities[:3]:
     if score > 0: 
        print(f"Documents {doc_index+1}: {document[doc_index]}")
```
